```python
# Importing the required libraries

import numpy as np
import pandas as pd
from scipy.stats import binom
```

```python
# Defining the parameters

n = 10 # The Number of trials
p = 0.5 # Probability of success on each trial (e.g. flipping a fair coin)
```

```python
# Generating binomial random variables

random_variable = binom.rvs(n, p, size = 1000)

# Generating 1000 random variables effectively means that 1000 instances of the binomial experiment should be simulated
# n represents the number of trials in each experiment
# p is the probability of success on each trial
```

```python
# Probability of getting exactly 5 heads out of 10 flips

prob_of_5_successes = binom.pmf(5,n,p)
print(" Probability of 5 successes :", prob_of_5_successes)
```

```python
# Probability of getting 5 or fewer successes

prob_up_to_5_successes = binom.cdf(5,n,p)
print(" Probability of 5 or fewer successes: ", prob_up_to_5_successes)
```

```python
# Plotting the Probability Mass distribution

import matplotlib.pyplot as plt

# Values of k to plot PMF for
k_values = np.arange(0, n+1) # Possible number of successes
pmf_values = binom.pmf(k_values, n, p)

# Create Plot
plt.bar(k_values, pmf_values)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.show()
```

```python
# Plotting the Cumulative Distribution Function

cdf_values = binom.cdf(k_values, n, p)

plt.step(k_values, cdf_values, where = 'mid', label = 'CDF')
plt.title('Binomial Distribution - CDF')
plt.xlabel('Number of Successes')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.legend()
plt.show()
```
