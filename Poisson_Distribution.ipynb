```python
# Importing the required libraries

import numpy as np
import pandas as pd
import math
import random
import math
import matplotlib.pyplot as plt
```

```python
# Random generation of 1000 independent Poisson samples

lambda_ = 7
N = 1000
X = np.random.poisson(lambda_, N)
```

```python
# Histogram of Poisson distribution

counts, bins, ignored = plt.hist(X, 50, density = True, color = 'purple')
plt.title("Randomly generate Poisson Distribution with lambda = 7")
plt.xlabel('Number of visitors')
plt.ylabel('Probability')
plt.show()
```

```python
from scipy.stats import poisson
```

```python
def poisson_probability (lambda_, k):
  return poisson.pmf(k, lambda_)
```

```python
def poisson_cumulative_probability(lambda_, k):
  return poisson.cdf(k, lambda_)
```

```python
if __name__ == "__main__":

  # Given parameters
  lambda_ = 10      #Average rate of lambda
  k = 5              #Number of events

  prob = poisson_probability(lambda_, k)
  print(f'Probability of observing {k} events: {prob:.4f}')
```
